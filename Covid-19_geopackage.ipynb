### Compilação do número de infetados por municipio com COVID-19 em formato espacial

#### Fonte dos dados:
- Carta Administrativa Oficial de Portugal (CAOP) de 2019 (DGT, 2020)
- Direção-Geral da Saúde . Relatório de Situação n.º 22 de 24/03/2020 (DGS, 2020)

```python
import geopandas
import matplotlib.pyplot as plt
```

```python
PATH_DATA = "../"
```

```python
#chamar o ficheiro do tipo csv
pt_conti = geopandas.read_file(f"{PATH_DATA}Geopackage_COVID.gpkg", layer='Municipios_Continente')
pt_madeira = geopandas.read_file(f"{PATH_DATA}Geopackage_COVID.gpkg", layer='ArqMadeira')
```

```python
print(pt_conti.head())
```

```python
pt_conti.plot(); pt_madeira.plot()
plt.show()
```

#### Apresentação dos valores disponibilizados em formato vetorial. Municipios com maior número  de infetados

```python
pt_conti.plot(column='co_2403', legend=True, figsize =[15, 9])
plt.show()
```

```python
pt_madeira.plot(column='co_2403', legend=True, figsize =[15, 10])
plt.show()
```
